```python
from modelicares import SimRes, SimResList
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import datetime
```

```python
data = SimResList('../Results/*year-occ*')
```

##Create the load duration curve

Prepare the data bij taken equidistant points

```python
load_names = [
    'hea.Q_flow',
    'buildingTest.heatingSystem.QHeaSys'
]

description = {
    'hea.Q_flow': {
        'label': 'Total load of the system',
        'unit': 'W'
    },
    'buildingTest.heatingSystem.QHeaSys': {
        'label': 'Space heating load',
        'unit': 'W'
    }
}

# Equi-distant time vector
minutes = 15
points = 31556926 / (60*minutes)
t = [60*minutes*i for i in range(24*(60/minutes), points)]
dates = map(datetime.datetime.fromtimestamp, t)
Te = []
for value in data['sim.Te'][0].values(t):
    Te.append(value._value - 273.15)    

# Get the loads at times t
loads = dict()
for name in load_names:
    loads[name] = data[name].values(t)
```

Plot the different metrics for each house

```python
for name in load_names:
    data.plot(name)
```

Create the load duration curve by summing the powers and sorting the array

```python
ldc = dict()
ldcs = dict()
for name in load_names:
    ldc[name] = [0 for i in range(24*(60/minutes), points)]
    ldcs[name] = [0 for i in range(24*(60/minutes), points)]
    
for name in load_names:
    for l in loads[name]:
        for i, q in enumerate(l):
            ldc[name][i] = ldc[name][i] + q._value
            ldcs[name][i] = ldc[name][i] + q._value
    ldcs[name].sort(reverse=True)

for key, value in ldcs.iteritems():
    ldcs[key] = np.asarray(value)
    belowzero = ldcs[key] < 0  
    ldcs[key][belowzero] = 0
    
for key, value in ldc.iteritems():
    ldc[key] = np.asarray(value)
    belowzero = ldc[key] < 0  
    ldc[key][belowzero] = 0

dhwl = np.sort(ldcs['hea.Q_flow']-ldcs['buildingTest.heatingSystem.QHeaSys'])[::-1]

plt.figure(figsize=(16,10))
for name in load_names:
    plt.plot(t[10:-1], ldcs[name][10:-1], label=description[name]['label'])

plt.plot(t, dhwl, label='DHW load')
    
plt.xlabel('Time')
plt.ylabel('Load [W]')
plt.ylim((0, 20000))
plt.legend()
```

## Reasons for the sharp peak:
- Bad sizing of the heating system (QNom too big) could lead to using all of the (exces) power at the same time step
- DHW profiles not scaled for our occupants

```python
for name in load_names:
    plt.figure(figsize=(16,10))
    plt.plot(t, ldc[name], label=description[name]['label'])
    plt.legend()
    plt.show()
```

```python
zoomstart = 86400*7
zoomstop = 86400*14

def plot_load(zoomstart, zoomstop):

    istart = zoomstart//(60*minutes)
    istop = zoomstop//(60*minutes)

    for name in load_names:
        plt.figure(figsize=(16,10))
        plt.plot(dates[istart:istop], ldc[name][istart:istop], label=description[name]['label'])
        plt.legend()
        plt.show()

    peaks = [
        18000,
        16000,
        14000,
        13000
    ]

    zoomdates = dates[istart:istop]

    fig, ax1 = plt.subplots()
    fig.set_size_inches(16,10)
    labels = []
    for i, peak in enumerate(peaks):
        index = ldc['hea.Q_flow'] > peak
        ax1.plot(zoomdates, index[istart:istop]*peak)
        labels.append(str(peak) + ' W')

    ax1.legend(labels=labels, loc=4)
    ax1.set_ylabel = 'Powers'
    ax1.set_ylim(0, 20000)

    ax2 = ax1.twinx()
    ax2.plot(zoomdates, Te[istart:istop], '-y')
    ax2.set_ylabel('Outside temperature', color='y')
    plt.legend(loc=8)
    plt.show()
```

```python
plot_load(0, -1)
```

```python
plot_load(86400*1, 86400*17)
```

```python
plot_load(-1*86400*7, -1)
```
