```python
from modelicares import SimRes
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
data = SimRes(r'../Results/SolarCollector.mat')
```

```python
Time = data['Time'].values()
Tin = data['temperature'].values()
mIn = data['massflow'].values()
Qsol = (data['solCol.solGai.HDirTil'].values() + data['solCol.solGai.HSkyDifTil'].values() + data['solCol.solGai.HGroDifTil'].values())* data['solCol.solGai.A_c'].values()[0]
Q = data['solCol.heaGai[1].Q_flow'].values() * data['solCol.nSeg'].values()[0]
```

```python
n = 2000
plt.plot(Time[0:n], data['solCol.solGai.HDirTil'].values()[0:n])
```

```python
plt.figure(figsize=(18,10))
n = 2000
plt.plot(Time[0:n], Qsol[0:n], Time[0:n], Q[0:n])
plt.legend(['Qsol', 'Q'])
```

```python
eff = Q[0:n]/Qsol[0:n]
for i, e in enumerate(eff):
    if e > 10:
        eff[i] = 0
        
plt.figure(figsize=(18,10))
plt.plot(Time[0:n],eff)
plt.title('efficiencies')
```

```python
zeroSolar = [i for i, q in enumerate(Q) if q > 10]
print len(zeroSolar)
print len(Qsol)
Tinstr = [Tin[z] for z in zeroSolar]
mInstr = [mIn[z] for z in zeroSolar]
Qsolstr = [Qsol[z] for z in zeroSolar]
Qstr = [Q[z] for z in zeroSolar]
```

```python
plt.plot(Qstr)
```

```python
import numpy as np
eff = np.array(Qstr)/np.array(Qsolstr)
        
plt.figure(figsize=(18,10))
plt.plot(eff)

# Remove extra-ordinary low efficiency values (probably due to transient behavior)
zeroSolar = [i for i, e in enumerate(eff) if e > 0.45]

Tinstr = [Tinstr[z] for z in zeroSolar]
mInstr = [mInstr[z] for z in zeroSolar]
Qsolstr = [Qsolstr[z] for z in zeroSolar]
Qstr = [Qstr[z] for z in zeroSolar]

eff = np.array(Qstr)/np.array(Qsolstr)
        
plt.figure(figsize=(18,10))
plt.plot(eff)
```

```python
import numpy as np
from scipy.stats import itemfreq

Tsteps = np.unique(Tinstr)
Msteps = np.unique(mInstr)

print 'Unique values for the massflow'
print '-------------------------------------------------------'
print itemfreq(mInstr)
print ''
print 'Unique values for the temperature'
print '-------------------------------------------------------'
print itemfreq(Tinstr)

# Find indexes of temperatures
idxs = [[i for i, t in enumerate(Tinstr) if t==Ts] for Ts in Tsteps]
# Count the different massflows for each temperature step


mt = {}
for j, idx in enumerate(idxs):
    m = [float(mInstr[i]) for i in idx]
    mt[Tsteps[j]] = {}
    for k, massflow in enumerate(Msteps):        
        mtIndexes = [l for l, value in enumerate(mInstr) if value==massflow and l in idx]
        mt[Tsteps[j]][Msteps[k]] = mtIndexes
```

```python
Ts = 7
Ms = 6

testIndex = mt[Tsteps[Ts]][Msteps[Ms]]
plt.plot([Qsolstr[i] for i in testIndex])
plt.show()
plt.plot([eff[i] for i in testIndex],[Qsolstr[i] for i in testIndex], 'o')
```

```python
for t in range(1):
    for m in range(5):
        testIndex = mt[Tsteps[t]][Msteps[m]]
        label = 'Temperature: '+ str(Tsteps[t]) + ' - Massflow: ' + str(Msteps[m])
        plt.plot([eff[i] for i in testIndex],[Qsolstr[i] for i in testIndex], 'o')
        plt.title(label)
        plt.show()
```

```python
from scipy.optimize import curve_fit

def f(x, a, b):
    return a*x + b

def f2(x, a, b, c):
    return a*x**2 + b*x + c

xdata = np.array([Qsolstr[i] for i in testIndex])
ydata = np.array([eff[i] for i in testIndex])

popt, pcov = curve_fit(f, xdata, ydata)
```

```python
plt.plot(xdata, ydata, 'o', xdata, f(xdata, popt[0], popt[1]))
```

```python
linearApprox = {}

for t in Tsteps:
    linearApprox[t] = {}
    for m in Msteps:
        linearApprox[t][m] = {}
        
        testIndex = mt[t][m]
        xdata = np.array([Qsolstr[i] for i in testIndex])
        ydata = np.array([eff[i] for i in testIndex])
        popt, pcov = curve_fit(f, xdata, ydata)
        
        linearApprox[t][m]['a'] = popt[0]
        linearApprox[t][m]['b'] = popt[1]
```

```python
# Test 5 random correlations

import random
for i in range(5):
    t = Tsteps[random.randint(0, len(Tsteps)-1)]
    m = Msteps[random.randint(0, len(Msteps)-1)]

    testIndex = mt[t][m]

    xdata = np.array([Qsolstr[i] for i in testIndex])
    ydata = np.array([eff[i] for i in testIndex])

    plt.plot(xdata, ydata, 'o', xdata, f(xdata, linearApprox[t][m]['a'], linearApprox[t][m]['b']))
    plt.show()
```

```python
# Save the a and b values to an excel file

import xlsxwriter

workbook = xlsxwriter.Workbook('data.xlsx')
worksheet = workbook.add_worksheet()

row = 2
col = 0

for i, string in enumerate(['T', 'm', 'a', 'b']):
    worksheet.write(1,i,string)


for key in linearApprox.keys():
    row += 1
    worksheet.write(row, col, key)
    for item in linearApprox[key]:
        worksheet.write(row, col + 1, item)
        worksheet.write(row, col + 2, linearApprox[key][item]['a'])
        worksheet.write(row, col + 3, linearApprox[key][item]['b'])
        
        row += 1

workbook.close()
```
