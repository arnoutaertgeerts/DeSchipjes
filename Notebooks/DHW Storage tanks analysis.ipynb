```python
import sys
sys.path.append('/home/arnout/Projects/Work/python-highcharts')

from modelicares import SimRes

import scipy as sp
import numpy as np
import charts
import os
import json

from charts.jsonencoder import ChartsJSONEncoder
```

```python
def to_json_files(sim, path, keys):
    if not os.path.exists(path):
        os.makedirs(path)

    dic = []
    for key in keys:
        v = sim[key]
        x = v.times().tolist()
        y = v.values().tolist()
        data = np.array([x,y]).T
        
        with open(os.path.join(path, key + '.json'), "w") as f:
            f.write(json.dumps(dict(name=key, data=data), cls=ChartsJSONEncoder))
        
        dic.append(dict(name=key, display=False))
    with open(os.path.join(path, 'keys.json'), "w") as f:
        f.write(json.dumps(dic))
        
def to_series(sim, key, name):
    dic = dict()
    
    v = sim[key]
    x = v.times().tolist()
    y = v.values().tolist()
    data = np.array([x,y]).T
    
    return dict(name=name, data=data)
```
