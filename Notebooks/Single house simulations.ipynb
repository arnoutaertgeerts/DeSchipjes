```python
import modelicares as mod
import numpy as np
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt

simulators = mod.simulators
```

```python
house1 = "../Dymosims/haarhakker"
house2 = "../Dymosims/peterslei"

results_dir = "Results"
StopTime=31556#926
```

Simulation settings dictionary

```python
settings = [dict() for i in range(11)]
```

Randomly choose 11 occupancy profiles

```python
from random import randint
for s in settings:
    s['buildingTest.occupant.id'] = randint(1,33)
```

```python
simulator = simulators.dymosim(results_dir=results_dir, StopTime=StopTime)
simulator_debug = simulators.dymosim(results_dir=results_dir, StopTime=StopTime, debug=True)
sims = []
```

Create the simulator and run the simulations

```python
print 'Starting simulations...'

for s in settings[:4]:
    sims.append(simulator.run(model=house1, params=s))
                
for s in settings[4:]:
    sims.append(simulator.run(model=house2, params=s))

for sim in sims:
    sim.wait()

for sim in sims:
    print 'Success:', sim.successful()
    
print 'Done!'
```
