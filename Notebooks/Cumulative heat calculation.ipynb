```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from modelicares import SimRes, SimResList

results = [
    SimRes(r'../Results/Building140.mat'),
    SimRes(r'../Results/Renovated.mat')
]
variables = ['building.heatingSystem.QHeaSys']
```

```python
sim = SimResList(results)
plt.figure(figsize=(15,10))
sim.plot(variables)
```

```python
cumulative = dict()
for r in results:
    cumulative[str(r)] = dict()
    for v in variables:
        cumulative[str(r)][v] = np.cumsum(r[v].values())
```

```python
i = 0
for rKey, rValue in cumulative.iteritems():
    for vKey, vValue in rValue.iteritems():
        plt.title('Cumulative sum')
        plt.plot(sim[i]['Time'].values()/(3600*24), vValue/1000)
        i += 1

plt.legend(['Old', 'Renovated'], loc=2)
plt.xlabel('Time / days')
plt.ylabel('kJ')
```
