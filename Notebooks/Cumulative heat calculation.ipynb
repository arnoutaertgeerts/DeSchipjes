```python
%pylab inline
from modelicares import SimRes

resultFile = 'C:\Users\u0098668\Documents\Dymola/ScenarioTest.mat'
variables = ['grid.building1.heatingSystem.QHeaSys', 'grid.building2.heatingSystem.QHeaSys']
```

```python
sim = SimRes(resultFile)
figsize(15,10)
sim.plot(variables)
```

```python
cumulative = {}
for v in variables:
    cumulative[v] = cumsum(sim[v].values())
```

```python
for key, value in cumulative.iteritems():
    title('Cumulative sum')
    plot(sim['Time'].values()/(3600*24), value/1000)

legend(variables)
xlabel('Time / days')
ylabel('kJ')
```
